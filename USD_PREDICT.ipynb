```python
import sklearn
```

```python
import pandas as pd
```

```python
money = pd.read_csv('USD_CURRENCY_HISTORY.csv', sep=";")
```

```python
past = 7 * 4  # берем данные за последние четыре недели
future = 7  # и по ним предсказываем на неделю вперед
```

```python
values = money["curs"].map(lambda x: float(str(x).replace(',', '.')))
```

```python
len(values)
```

```python
start = past
end = len(values) - future
raw_df = []

for i in range(start, end):
    past_and_future_value = values[(i-past):(i+future)]
    raw_df.append(list(past_and_future_value))
    
```

```python
df
```

```python
past_columns = [f"past_{i}" for i in range(past)]
future_columns = [f"future_{i}" for i in range(future)]
```

```python
df = pd.DataFrame(raw_df, columns=(past_columns + future_columns))
```

```python
df.head(5)
```

```python
X = df[past_columns][:-1] # это данные на основе, которых делаются предсказания
y = df[future_columns][:-1] # это собственно данные, которые будем предсказывать
```

```python
X_test = df[past_columns][-1:]
y_test = df[future_columns][-1:] # тестовые данные для проверки модели
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LinReg = LinearRegression() # здесь можно настроить модель
```

```python
LinReg.fit(X, y) # обучение модели
```

```python
LinReg.predict(X_test)
```

```python
y_test
```

```python
prediction = LinReg.predict(X_test)[0]
```

```python
import numpy as np
import sklearn
```

```python
np.array([prediction]).shape
np.array([list(y_test.iloc[0])]).shape
```

```python
sklearn.metrics.mean_absolute_error(np.array([list(y_test.iloc[0])]), np.array([prediction]))
```
